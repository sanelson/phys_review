To start with, spheres 1 and 2 each have charge q and experience a mutual repulsive force:

$$ F = \frac{k q^2}{r^2} $$

1. When sphere 3 touches sphere 1, sphere 1's charge decreases to q/2.
2. Then when sphere 3 (now carrying charge q/2) touches sphere 2 a total amount of q/2 + q becomes shared equally between them.
3. Therefore, the charge of sphere 3 is 3q/4 in the final state.

The repulsive force between the spheres:

$$ F^\prime = k \frac{(q/2)(3q/4)}{r^2} = \frac{3}{8}k\frac{q^2}{r^2} = \frac{3}{8}F 
\Rightarrow \frac{F^\prime}{F} = \frac{3}{8} = 0.375$$


## Problem 3
***
What must be the distance between point charge $q_1 = 26.0\ \mu\mathrm{C}$ and point charge $q_2 = 47.0\ \mu\textrm{C}$ for the electrostatic force between them to have a magnitude of 5.70 N?

## Solution

Given the the known quantities, we must solve for the distance $r$.

$$ F = k\frac{q_1 q_2}{r^2} 
    \quad \Rightarrow \quad r^2 = k\frac{q_1 q_2}{F} 
    \quad \Rightarrow \quad r = \pm\sqrt{k\frac{q_1 q_2}{F}}
$$

Inputing values (considering only positive distance $r$):

```python
q_1 = 26.0 * u.uC
q_2 = 47.0 * u.uC
F = 5.70 * u.N

#r = sqrt(1/(4*pi*eps0)*(q_1*q_2)/F)
r = np.sqrt(1/(4*pi*eps0)*(q_1*q_2)/F)
#r.base * r.unit.compose()[0]
r.to(u.m)
```

## Problem 4
***
In the return stroke of a typical lightning bolt, a current of $2.5\times10^4\ \textrm{A}$ exists for $20\ \mu\textrm{s}$. How much charge is transferred in this event?

## Solution

Current in *Amperes* is defined as:

$$ i = \frac{dq}{dt} $$

We want to find the amount of charge transfered or the change in charge for this event. Lets solve for $dq$.

$$ dq = i \cdot dt $$

Plug in values:

$$ dq = (2.5 \times 10^4\ \textrm{A})(20\ \mu\textrm{s}) $$

```python
i = 2.5e4 * u.A
t = 20 * u.us

q = i*t
q.to(u.C)
```

## Problem 5
***
A particle of charge $+3.00 \times 10^{-6}\ \textrm{C}$ is 12.0 cm distant from a second particle of charge $-1.50 \times 10^{-6}\ \textrm{C}$. Calculate the magnitude of the electrostatic force between the particles.

## Solution

Plug in values:

$$ F = k\frac{q_1 q_2}{r^2} = \frac{1}{4\pi\epsilon_0} \frac{|q_1||q_2|}{r^2}$$

$$ F = (8.99 \times 10^9\ \textrm{N}\cdot\textrm{m}^2/\textrm{C}^2)\times
       \frac{|+3.00 \times 10^{-6}\ \textrm{C}||-1.50 \times 10^{-6}\ \textrm{C}|}
           {12.0\ \textrm{cm}}
$$

```python
#k = 1/(4*pi*eps0)
#k.base * k.unit.compose()[0]
```

```python
q_1 = 3.00e-6 * u.C
q_2 = -1.5e-6 * u.C
r = 12.0 * u.cm

F =  1/(4*pi*eps0) * (abs(q_1)*abs(q_2))/r**2
F.to(u.N)
```

## Problem 6
***
Two equally charged particles are held $3.2 \times 10^{-3}\ \textrm{m}$ apart and then released from rest. The initial acceleration of the first particle is observed to be $7.0\ \textrm{m}/\textrm{s}^2$ and that of the second to be $9.0\ \textrm{m}/\textrm{s}^2$. If the mass of the first particle is $6.3 \times 10^{-7}\ \textrm{kg}$, what are (a) the mass of the second particle and (b) the magnitude of the charge of each particle?

## Solution

Given the charges $q$ are equal, the particles will be moving away from each other after release. We can also note that the electrostatic force exerted by each particle will be equal, since that does not really depend on mass, only charge and distance.

$$ q_1 = q_2 = q $$

$$ F_1 = F_2 = F \quad \Rightarrow \quad F = m_1 a_1 = m_2 a_2 $$

Solve for the mass of the second particle:

$$ \frac{m_1 a_1}{a_2} = m_2 $$

```python
# Part (a)
m_1 = 6.3e-7 * u.kg
a_1 = 7.0 * u.m / u.s**2
a_2 = 9.0 * u.m / u.s**2

m_2 = m_1*a_1/a_2
ans('a', m_2)
```

The magnitude of charge can be calculated from the force, charge and distance.

Force is $F = m_1 a_1$



```python
F = m_1 * a_1
F.to(u.N)
```

Now, solve Coulomb's law for the charge $q$.

$$ F = \frac{1}{4\pi\epsilon_0} \frac{q^2}{r^2} 
     \quad \Rightarrow \quad
     F r^2 4\pi\epsilon_0 = q^2
     \quad \Rightarrow \quad
     q = \pm\sqrt{F r^2 4\pi\epsilon_0}
$$

Since we are looking for the magnitude of charge, only worry about positive values of $q$

```python
# Part b
r = 3.2e-3 * u.m

q = np.sqrt(F*4*pi*eps0*r**2)
ans('b', q.to(u.C))
```
